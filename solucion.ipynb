```python
# Paso 1: Importar las librerías necesarias
import pandas as pd
import numpy as np
```

```python
# Paso 2: Cargar los datos desde los archivos CSV
# Cargar los datos con rutas absolutas
ventas = pd.read_csv('/Users/adriana/Master/R Sutdio/tarea_python/template/proyecto_final_pandas/ventas.csv')
inventarios = pd.read_csv('/Users/adriana/Master/R Sutdio/tarea_python/template/proyecto_final_pandas/inventarios.csv')
satisfaccion = pd.read_csv('/Users/adriana/Master/R Sutdio/tarea_python/template/proyecto_final_pandas/satisfaccion.csv')
```

```python
# Paso 3: Limpieza de datos
# Eliminar filas con valores nulos

print('Na para ventas\n',ventas.isnull().sum())
print('Na para inventarios\n',inventarios.isnull().sum())
print('Na para satisfacción\n',satisfaccion.isnull().sum())

ventas.dropna(inplace=True)
inventarios.dropna(inplace=True)
satisfaccion.dropna(inplace=True)
```

```python
#  Exploración de datos (Pandas) 
#Calcula las ventas totales por producto y por tienda.
ventas['Ingresos'] = ventas['Precio_Unitario'] * ventas['Cantidad_Vendida'] #Añadimos la columna ingresos al data set
print(ventas)
ventas_por_tienda = ventas.groupby('ID_Tienda')['Ingresos'].sum() #Agrupamos y sumamos por tienda
ventas_por_producto = ventas.groupby('Producto')['Ingresos'].sum() #Agrupamos y susmamos por producto
print('Total de ventas por tienda\n',ventas_por_tienda)
print('Total de ventas por producto\n',ventas_por_producto)
```

```python
#Calcular la rotación de inventarios por tienda.
union = ventas.merge(inventarios, on= ['ID_Tienda','Producto']) #Unir los datos de ventas e inventario

union['Rotacion'] = (union['Cantidad_Vendida']/ union['Stock_Disponible'])*100
print(union)
```

```python
#Filtrar las tiendas con inventarios críticos (menos del 10% en ventas respecto al inventario disponible).
baja_rotacion = union[union['Rotacion'] < 10]
print("Los productos con baja rotación son:", baja_rotacion)
```

```python
#Analizar la satisfacción del cliente y filtrar las tiendas con una satisfacción menor al 60%.
baja_satisfaccion = satisfaccion[satisfaccion['Satisfacción_Promedio'] < 60]
print('Las tiendas con satisfacción inferior al 60% son:', baja_satisfaccion)
```

```python
#Calcular la mediana de las ventas totales.
mediana_ventas = union['Ingresos'].median()
print('La mediana de ventas es: ',mediana_ventas)

#Calcular la desviación estándar de las ventas.
desviacion_estandar =  union['Ingresos'].std()
print('La desviación estandar es de: ', desviacion_estandar)
```
